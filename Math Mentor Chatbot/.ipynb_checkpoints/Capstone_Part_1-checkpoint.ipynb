```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import requests
```

```python
driver = webdriver.Safari()
driver.get('https://ischool.sg/Topics/FRACTIONS-122/')

try:
    # Wait for the presence of the elements before finding them
    text_elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'question-item-content_full'))
    )

    # Iterate over each text element and print its text
    for i, element in enumerate(text_elements):
        print(f"Text from element {i+1}: {element.text}")

    # Find and download images as before
    img_elements = driver.find_elements(By.CSS_SELECTOR, '.col-xs-12.col-sm-2.row img')
    image_counter = 1

    for img_element in img_elements:
        img_url = img_element.get_attribute('src')
        if img_url:
            response = requests.get(img_url)
            if response.status_code == 200:
                filename = f'downloaded_image_{image_counter}.jpg'
                with open(filename, 'wb') as file:
                    file.write(response.content)
                print(f"Image {image_counter} downloaded successfully.")
                image_counter += 1
            else:
                print(f"Failed to download image {image_counter}, HTTP status code: {response.status_code}")
        else:
            print(f"Image URL for image {image_counter} not found")

finally:
    driver.quit()
```

```python
# Re-importing necessary libraries and reloading the data since the execution state was reset
import pandas as pd
import json
import re
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Function definitions
def avg_word_length(text):
    words = text.split()
    average = sum(len(word) for word in words) / len(words) if words else 0
    return average

def avg_sentence_length(text):
    sentences = re.split(r'[.!?]', text)
    sentences = [s.strip() for s in sentences if s.strip()]
    average = sum(len(s.split()) for s in sentences) / len(sentences) if sentences else 0
    return average

# Reloading the data
file_path = 'test.json'
with open(file_path, 'r') as file:
    data = json.load(file)

df = pd.DataFrame(data)
df['avg_word_length'] = df['Problem'].apply(avg_word_length)
df['avg_sentence_length'] = df['Problem'].apply(avg_sentence_length)
df['question_length'] = df['Problem'].apply(len)

# Selecting and standardizing the features
features = df[['question_length', 'avg_word_length', 'avg_sentence_length']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Applying K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(features_scaled)

# Displaying the first few rows with the cluster assignments
df[['Problem', 'question_length', 'avg_word_length', 'avg_sentence_length', 'cluster']].head(30)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA
import pandas as pd

# Load your data
# Assuming 'questions' is a list of question strings
df = pd.read_csv('clean.csv')
questions = df['Stem'].tolist()

# Preprocess and vectorize the text data
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(questions)

# Normalize the TF-IDF vectors
tfidf_norm = normalize(tfidf_matrix)

# Reduce dimensions (optional, can help if the number of features is very high)
pca = PCA(n_components=0.95)  # retain 95% variance
tfidf_pca = pca.fit_transform(tfidf_norm.toarray())

# Choose the number of clusters (k) using the elbow method or other approach
# For the sake of this example, let's say we choose 5 clusters
num_clusters = 5
km = KMeans(n_clusters=num_clusters)

# Fit the K-means model
km.fit(tfidf_pca)

# Get the cluster assignments for each question
clusters = km.labels_.tolist()

# Assign the cluster labels to the original dataframe
df['Cluster'] = clusters

# Now df has an additional column 'Cluster' that indicates the assigned cluster for each question
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load your questions dataset
df = pd.read_csv('clean.csv')
questions = df['Stem'].tolist()

# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')  # A commonly used model for generating embeddings

# Generate embeddings for all questions
question_embeddings = model.encode(questions)

# To find similar questions for a given query question
query_question = "What is the result of dividing 20 by 4?"
query_embedding = model.encode([query_question])

# Compute similarities between the query and all question embeddings
similarities = cosine_similarity(query_embedding, question_embeddings)

# Find the index of the most similar question
most_similar_index = similarities[0].argmax()
most_similar_question = questions[most_similar_index]

print("The most similar question to the query is:", most_similar_question)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load your questions dataset
df = pd.read_csv('clean.csv')
questions = df['Stem'].tolist()

# Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for all questions
question_embeddings = model.encode(questions)

# Your query question
query_question = "Find the sum of 8 and 5."
query_embedding = model.encode([query_question])

# Compute cosine similarities
similarities = cosine_similarity(query_embedding, question_embeddings)

# Get a list of scores and corresponding questions
similarity_scores = similarities[0]
questions_similarity = list(zip(questions, similarity_scores))

# Sort by similarity scores in descending order
sorted_similar_questions = sorted(questions_similarity, key=lambda x: x[1], reverse=True)

# Take the top 5 most similar questions
top_5_similar_questions = sorted_similar_questions[1:6]  # Start from 1 to skip the query itself

# Print out the top 5 similar questions
for question, score in top_5_similar_questions:
    print(f"Question: {question} - Similarity Score: {score}")
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from sklearn.model_selection import train_test_split
from datasets import Dataset

# Load the labeled data
df_labeled = pd.read_csv('labelled.csv')

# Let's say your dataframe has two columns: 'question' and 'difficulty'
texts = df_labeled['Stem'].tolist()
labels = df_labeled['Label'].tolist()

# Convert labels to numeric
label_dict = {label: idx for idx, label in enumerate(set(labels))}
labels = [label_dict[label] for label in labels]

# Split the data into a training and a validation set
train_texts, val_texts, train_labels, val_labels = train_test_split(texts, labels, test_size=0.2)

# Tokenize the text
tokenizer = AutoTokenizer.from_pretrained("tbs17/MathBERT")
train_encodings = tokenizer(train_texts, truncation=True, padding=True)
val_encodings = tokenizer(val_texts, truncation=True, padding=True)

# Convert to PyTorch datasets
train_dataset = Dataset.from_dict({'input_ids': train_encodings['input_ids'], 'attention_mask': train_encodings['attention_mask'], 'labels': train_labels})
val_dataset = Dataset.from_dict({'input_ids': val_encodings['input_ids'], 'attention_mask': val_encodings['attention_mask'], 'labels': val_labels})

# Load the MathBERT model with a classification head
model = AutoModelForSequenceClassification.from_pretrained("tbs17/MathBERT", num_labels=len(label_dict))

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset
)

# Train the model
trainer.train()

# Evaluate the model
trainer.evaluate()
```

```python
import torch
from torch.utils.data import DataLoader
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Make sure to import pandas if not already done
import pandas as pd

# Assuming df_unlabeled is your DataFrame with unlabeled data and the questions are in the 'Stem' column
df_unlabeled = pd.read_csv('unlabelled.csv')
questions = df_unlabeled['Stem'].tolist()

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("tbs17/MathBERT")
model = AutoModelForSequenceClassification.from_pretrained("./results")  # Adjust the path to your model

# Tokenize the questions
encodings = tokenizer(questions, truncation=True, padding=True, return_tensors="pt", max_length=512)  # Specify max_length if needed

# Create a DataLoader
dataset = torch.utils.data.TensorDataset(encodings['input_ids'], encodings['attention_mask'])
data_loader = DataLoader(dataset, batch_size=16)

# Predict
predictions = []
for batch in data_loader:
    with torch.no_grad():
        inputs = {'input_ids': batch[0], 'attention_mask': batch[1]}
        outputs = model(**inputs)
        logits = outputs.logits
        predictions.extend(torch.argmax(logits, dim=1).tolist())

# Map predictions back to labels
id_to_label = {idx: label for label, idx in label_dict.items()}
predicted_labels = [id_to_label[pred] for pred in predictions]

# Add predictions to the DataFrame
df_unlabeled['PredictedLabel'] = predicted_labels

# Save or use df_unlabeled as needed
```
