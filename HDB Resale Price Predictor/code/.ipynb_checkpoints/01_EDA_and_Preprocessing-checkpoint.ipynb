### Problem Statement:
The purpose of the HDB is is egalitarian - to provide affordable housing to all Singaporeans. However, certain areas seem to be much more expensive than others. This seems to run counter to the HDB's goal. We wish to further the HDB's goals by applying an equalising pressure on both sides of the price spectrum. Therefore, we intend to use data to determine what amenities cause the highest increase in price in order for the efficient deployment of state resources in bringing up the desirability of seemingly undervalued units.

```python
#function to find the value counts for a particular dtype and save into json for easy visualization

def value_counts_dtype(name,df):
    
    value_counts_dict = {}

    for column in df.columns:
        value_counts = df[column].value_counts().sort_values(ascending=False).to_dict()
        value_counts_dict[column] = value_counts

    with open(f'../reports/{name}_value_counts.json', 'w') as json_file:
        json.dump(value_counts_dict, json_file, indent=4)
```

```python
#finding value counts for different dtypes
value_counts_dtype('object',object_df)
value_counts_dtype('integer',integer_df)
value_counts_dtype('float',float_df)
```

```python
#replace the Nan values with 0

missing_cols = ['mall_within_2km','mall_within_1km','mall_within_500m',
                'hawker_within_2km','hawker_within_1km','hawker_within_500m']

for col in missing_cols:
    df[col].fillna(value=0,inplace=True)
```

```python
#check that there are no more nan values

print('Count of NAN values')
print('-------------------')
for col in missing_cols:
    print(col,':\t', df[col].isna().sum())
```

```python
duplicates = []
cols = df.columns
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        if df[cols[i]].equals(df[cols[j]]):
            duplicates.append((cols[i], cols[j]))

print("Duplicate columns:", duplicates)
```

```python
#impute
from sklearn.impute import SimpleImputer

#chose median since data is right skewed and median is not affected by outliers
simple_imputer = SimpleImputer(strategy='median') 
simple_imputer.fit(df[['mall_nearest_distance']])
imputed_column = pd.DataFrame(simple_imputer.transform(df[['mall_nearest_distance']]))
df['mall_nearest_distance'] = imputed_column
```
