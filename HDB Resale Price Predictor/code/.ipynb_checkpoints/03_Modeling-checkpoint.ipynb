## Part 3: Modeling

```python
import pandas as pd
import time
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures, MinMaxScaler, RobustScaler, PowerTransformer
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split, KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RidgeCV, ElasticNet
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
%%time

df = pd.read_csv('../datasets/clean_train.csv', low_memory=False)
df.columns

features=["town", "floor_area_sqm", "mid", "hdb_age", "market_hawker", 
          "commercial", "multistorey_carpark", "mall_nearest_distance", 
          "hawker_food_stalls", "hawker_market_stalls", "mrt_nearest_distance", 
          "pri_sch_nearest_distance", "pri_sch_name", "sec_sch_name", "tranc_year", 
          "flat_type", "flat_model","max_floor_lvl","planning_area","cutoff_point"]

numeric_cols = df.select_dtypes(include=[np.number])

z_scores = np.abs(stats.zscore(numeric_cols))
threshold = 3

df = df[(z_scores < threshold).all(axis=1)]

X = df[features]
y = df['resale_price']

num_cols = X.select_dtypes(include=['float64', 'int64']).columns
cat_cols = X.select_dtypes(include=['object']).columns

# mall_distance_col = ['mall_nearest_distance']
# other_num_cols = [col for col in num_cols if col != 'mall_nearest_distance']

distance_cols = ['mall_nearest_distance', 'mrt_nearest_distance', 'pri_sch_nearest_distance']
other_num_cols = [col for col in num_cols if col not in distance_cols]

distance_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', PowerTransformer())
])

# mall_distance_pipeline = Pipeline([
#     ('imputer', SimpleImputer(strategy='median')),
#     ('scaler', StandardScaler())
# ])

other_num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value=0)),
    ('scaler', MinMaxScaler())
])

cat_pipeline = Pipeline([
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('distance', distance_pipeline, distance_cols),
        ('other_num', other_num_pipeline, other_num_cols),
        ('cat', cat_pipeline, cat_cols)
    ])

# preprocessor = ColumnTransformer(
#     transformers=[
#         ('mall_distance', mall_distance_pipeline, mall_distance_col),
#         ('other_num', other_num_pipeline, other_num_cols),
#         ('cat', cat_pipeline, cat_cols)
#     ])

linear_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

ridge_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('ridge', Ridge())
])

ridgecv_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('ridgecv', RidgeCV(alphas=[0.01, 0.1, 1, 10, 100], store_cv_values=True))
])

ridge_poly_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('poly', PolynomialFeatures(degree=2)),
    ('ridgecv_poly', Ridge())
])

elastic_net_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('elastic_net', ElasticNet())
])

linear_param_grid = {
    'regressor__fit_intercept': [True, False]
}

ridge_param_grid = {
    'ridge__alpha': [0.01, 0.1, 1, 10, 100]
}

elastic_net_param_grid = {
    'elastic_net__alpha': [0.01, 0.1, 1, 10, 100],
    'elastic_net__l1_ratio': [0.1, 0.5, 0.7, 0.9, 1]
}

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

pipelines = [linear_pipeline, ridge_pipeline, ridgecv_pipeline, ridge_poly_pipeline, elastic_net_pipeline]
param_grids = [linear_param_grid, ridge_param_grid, {}, {}, elastic_net_param_grid]

for pipe, params in zip(pipelines, param_grids):
    grid_search = GridSearchCV(pipe, param_grid=params, cv=KFold(n_splits=8), scoring='neg_root_mean_squared_error', verbose=1)
    grid_search.fit(X_train, y_train)
    print(f"Best parameters for {pipe.steps[-1][0]}: {grid_search.best_params_}")
    print(f"Best score for {pipe.steps[-1][0]}: {-grid_search.best_score_}")
```

```python
model = grid_search.best_estimator_
y_pred = model.predict(X_test)
```

```python
rmse_train = mean_squared_error(y_train, model.predict(X_train), squared=False)
rmse_test = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error (Train):", rmse_train)
print("Root Mean Squared Error (Test):", rmse_test)
```

```python
mae_train = mean_absolute_error(y_train, model.predict(X_train))
mae_test = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (Train):", mae_train)
print("Mean Absolute Error (Test):", mae_test)
```

```python
r2_train = model.score(X_train, y_train)
r2_test = model.score(X_test, y_test)
print("R2 Score (Train):", r2_train)
print("R2 Score (Test):", r2_test)
```
